```python
# Import libraries
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time

# Set up Chrome WebDriver
chrome_options = Options()
chrome_options.add_argument('--headless')  # To run Chrome in headless mode
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_experimental_option('excludeSwitches', ['enable-logging'])  # Exclude logging
chrome_options.add_experimental_option('useAutomationExtension', False)  # Disable usage data sending
driver = webdriver.Chrome(options=chrome_options)

# Function to scrape data from a given URL
def scrape_data(url):
    driver.get(url)
    time.sleep(3)  # Allow time for dynamic content to load

    # Get the HTML content
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Find all doctor listings
    doctor_listings = soup.find_all('div', class_='dirlist_box list-page')

    data = []
    for listing in doctor_listings:
        name = listing.find('h2').text.strip()
        specialty = listing.find('small').text.strip()
        location = listing.find('span', class_='capitalize').text.strip()

        data.append([name, specialty, location])

    return data

# Base URL of the website with multiple pages
base_url = 'https://www.medindia.net/patients/doctor_search/doctors-bangalore-{}.htm'

# List to store all scraped data
all_data = []

# Iterate through each page (up to page 700)
for page_num in range(1, 701):  # Adjust the range as needed
    url = base_url.format(page_num)
    data = scrape_data(url)
    all_data.extend(data)

# Create a DataFrame
df = pd.DataFrame(all_data, columns=['Name', 'Specialty', 'Location'])

# Export DataFrame to CSV
df.to_csv('doctors_info_full.csv', index=False)

# Quit WebDriver
driver.quit()
```

```python
import hashlib

# Function to generate unique code from name
def generate_code(name):
    # Using MD5 hashing algorithm to create a unique code
    return hashlib.md5(name.encode()).hexdigest()

# Generate unique codes for names
df['Code'] = df['Name'].apply(generate_code)

# Drop duplicates based on unique code
df.drop_duplicates(subset='Code', keep='first', inplace=True)

# Export DataFrame to CSV
df.to_csv('doctors_info_unique.csv', index=False)
```

```python
import json

def merge_json(file1, file2, merged_file):
    # Read data from the first JSON file
    with open(file1, 'r') as f:
        data1 = json.load(f)

    # Read data from the second JSON file
    with open(file2, 'r') as f:
        data2 = json.load(f)

    # Concatenate the lists from both JSON files
    merged_data = data1 + data2

    # Write the merged data into a new JSON file
    with open(merged_file, 'w') as f:
        json.dump(merged_data, f, indent=4)

# Paths to the JSON files
file1_path = '/content/doctors_info_unique_with_url.json'
file2_path = '/content/doctors_info_unique_with_url2.json'
merged_file_path = 'merged_file.json'

# Merge the JSON files
merge_json(file1_path, file2_path, merged_file_path)

print("JSON files merged successfully!")
```
